####  Basics of Topic Modeling


**Introduction to Topic Modeling Concepts**

Topic modeling is a machine learning technique used to uncover hidden themes or topics in a collection of documents. It helps in identifying patterns in the text data, making it easier to organize, search, and understand large volumes of information.

**Traditional Methods: LDA and NMF**

- **Latent Dirichlet Allocation (LDA)**:
  - **Concept**: LDA is a generative probabilistic model. It assumes each document is a mixture of a small number of topics, and each word in the document is attributable to one of the document's topics.
  - **How it works**: LDA identifies patterns in the distribution of words across documents. It uses these patterns to assign words to topics and documents to topic mixtures.
  - **Example**: If you have a set of news articles, LDA might identify topics like politics, sports, and technology based on the word distribution.

- **Non-Negative Matrix Factorization (NMF)**:
  - **Concept**: NMF is a linear algebra-based method. It factorizes the document-term matrix into two lower-dimensional matrices, representing topics and topic compositions.
  - **How it works**: NMF decomposes the original matrix into non-negative factors, ensuring the resulting matrices are interpretable and aligned with the actual topics.
  - **Example**: For a set of scientific papers, NMF could identify topics like biology, chemistry, and physics by factorizing the frequency of words used in the documents.

**Advantages and Limitations of Traditional Methods**

- **Advantages**:
  - **LDA**:
    - Well-established and widely used.
    - Provides a clear probabilistic framework.
  - **NMF**:
    - Produces easily interpretable results.
    - Effective for smaller datasets and simpler models.

- **Limitations**:
  - **LDA**:
    - Struggles with large and complex datasets.
    - Requires careful tuning of hyperparameters.
  - **NMF**:
    - Less effective for highly varied and large datasets.
    - Can produce less stable results compared to probabilistic models.

### How BERTopic Works: Overview of BERT Embeddings and Clustering


#### BERT Embeddings

**BERT** (Bidirectional Encoder Representations from Transformers): A powerful deep learning model. 

1. **Contextual Understanding**: BERT can understand the meaning of a word based on the words around it. This is different from traditional models that assign a fixed meaning to each word.
   - **Example**: In "I went to the bank to deposit money" and "I sat on the river bank," the word "bank" has different meanings. BERT uses the surrounding words to understand that one "bank" is a financial institution and the other is the side of a river.


````python
from transformers import BertTokenizer, BertModel
import warnings
import torch


# Load the pre-trained model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Sentences to analyze
sentences = [
    'I went to the bank to deposit money.',
    'I sat on the river bank.'
]

# Tokenize the sentences and convert into input IDs
inputs = tokenizer(sentences, return_tensors='pt', padding=True, truncation=True)

# Get the embeddings from BERT
with torch.no_grad():
    outputs = model(**inputs)

# Extract the embeddings for the word 'bank' in each sentence
# 'bank' appears at index 5 in the first sentence and index 5 in the second sentence
embedding_bank_1 = outputs.last_hidden_state[0, 5, :]  # Extracts the embedding for 'bank' from the first sentence
embedding_bank_2 = outputs.last_hidden_state[1, 5, :]  # Extracts the embedding for 'bank' from the second sentence

# Compare the embeddings
similarity = torch.cosine_similarity(embedding_bank_1.unsqueeze(0), embedding_bank_2.unsqueeze(0))

print(f"Cosine similarity between 'bank' in the two sentences: {similarity.item()}")
````

2. **Embeddings**: BERT converts words into vectors, which are numerical representations capturing the context and meaning of the words. This allows BERT to understand the nuances and relationships between words in a given sentence.

````python
from transformers import BertTokenizer, BertModel
import torch

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Sentence to analyze
sentence = "The bank will not accept the deposit after 3 PM."

# Tokenize the sentence and convert to input IDs
inputs = tokenizer(sentence, return_tensors='pt')

# Get the embeddings from BERT
with torch.no_grad():
    outputs = model(**inputs)

# Extract the embeddings for all tokens in the sentence
embeddings = outputs.last_hidden_state

# Print the shape of the embeddings tensor
print(embeddings.shape)

# Get the tokenized input IDs
input_ids = inputs['input_ids'][0]

# Find the index of the word 'bank' in the tokenized input
token_index = (input_ids == tokenizer.convert_tokens_to_ids('bank')).nonzero(as_tuple=True)[0].item()

# Extract and print the embedding for the word 'bank'
embedding_bank = embeddings[0, token_index, :]
print(embedding_bank)
````

#### Clustering

After BERT has created these embeddings, BERTopic clusters them. Clustering means grouping similar texts together. Hereâ€™s how it works:

1. **HDBSCAN (Hierarchical Density-Based Spatial Clustering of Applications with Noise)** is a clustering algorithm used in BERTopic. Here's a simple breakdown:

* **Handles Noise:** It can deal with noisy data, meaning it can ignore irrelevant or random data points.
* **Varying Densities:** It can find clusters of different sizes and densities, making it versatile for real-world data.


````python
#Code Example HDBSCAN 
import hdbscan
import numpy as np
import matplotlib.pyplot as plt

# Sample data: 2D points
data = np.array([
    [1, 2], [2, 3], [3, 2], [8, 7], [8, 8], [25, 80],
    [24, 79], [26, 81], [50, 50], [51, 51], [52, 52]
])

# Create and fit HDBSCAN model
clusterer = hdbscan.HDBSCAN(min_cluster_size=2)
cluster_labels = clusterer.fit_predict(data)

# Plotting the clusters
plt.scatter(data[:, 0], data[:, 1], c=cluster_labels, cmap='rainbow')
plt.title('HDBSCAN Clustering')
plt.show()
````

2. **Grouping**: Similar texts get grouped into topics.
   - **Example**: Imagine you have a bunch of product reviews. HDBSCAN can group reviews mentioning "battery life" into one cluster, "design" into another, and "performance" into another. This way, you get clear topics from a large set of texts.


````python
#Example Grouping
import hdbscan
from sklearn.feature_extraction.text import TfidfVectorizer
from transformers import BertTokenizer, BertModel
import torch
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
# Sample product reviews
reviews = [
    "The battery life of this product is amazing.",
    "I love the design and build quality.",
    "Performance is excellent, very fast.",
    "Battery life could be better.",
    "The design is sleek and modern.",
    "Great performance for gaming and work.",
    "Battery lasts all day with heavy use.",
    "The build quality feels premium.",
    "The device heats up quickly during use.",
    "Elegant and stylish design.",
    "Superb gaming performance, no lag at all.",
    "Battery drains quickly when using high-power apps.",
    "Beautiful design, very aesthetic.",
    "Handles multitasking with ease.",
    "Battery performance exceeds expectations.",
    "The design stands out in the crowd.",
    "Runs multiple apps smoothly.",
    "Battery needs frequent charging.",
    "Chic and modern look.",
    "Excellent speed and performance under load."
]

# Step 1: Convert reviews to vectors using TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(reviews)

# Step 2: Apply HDBSCAN clustering
clusterer = hdbscan.HDBSCAN(min_cluster_size=2)
cluster_labels = clusterer.fit_predict(X.toarray())

# Print the clusters
for label in set(cluster_labels):
    print(f"Cluster {label}:")
    for i, review in enumerate(reviews):
        if cluster_labels[i] == label:
            print(f"  - {review}")
````

````python
pca = PCA(n_components=2)
reduced_X = pca.fit_transform(X.toarray())

plt.scatter(reduced_X[:, 0], reduced_X[:, 1], c=cluster_labels, cmap='rainbow')
plt.title('HDBSCAN Clustering')
plt.show()
````

### Hierarchical Clustering: 

Hierarchical clustering creates a tree-like structure of clusters. We'll focus on agglomerative (bottom-up) clustering, where each data point starts as its own cluster and merges step-by-step.

* 1. **Agglomerative Bottom-Up Approach**
* Concept: Starts with each data point as an individual cluster. Merges the closest pairs of clusters step-by-step.

* 2. **Merging Clusters**

* Concept: As clusters merge, you can see how individual data points form larger clusters. The dendrogram visually represents this process.


````python
#Example Agglomerative Approach

import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Sample data
data = np.array([
    [1, 2],
    [2, 3],
    [3, 3],
    [8, 7],
    [8, 8],
    [25, 80]
])

# Perform hierarchical/agglomerative clustering
linked = linkage(data, method='single')

# Create a dendrogram
plt.figure(figsize=(4, 4))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram - Bottom-Up Approach')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
````

````python
#Example Merging Cluster
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Sample data
data = np.array([
    [1, 2],
    [2, 3],
    [3, 3],
    [8, 7],
    [8, 8],
    [25, 80]
])

# Perform hierarchical/agglomerative clustering using complete linkage method
linked = linkage(data, method='complete')

# Create a dendrogram
plt.figure(figsize=(4,4))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram - Merging Clusters')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
````

### Role of UMAP in BERTopic

Let's break down how UMAP helps in BERTopic.

**UMAP (Uniform Manifold Approximation and Projection)** is a technique that reduces the number of dimensions in your data. Hereâ€™s why this is important for BERTopic:

#### Dimensionality Reduction

1. **High-Dimensional Data**: BERT embeddings are often very high-dimensional, meaning each word or sentence can be represented in a space with many dimensions (e.g., 768 dimensions).
   - **Example**: Imagine you have a spreadsheet with 768 columns. Each column represents a different aspect of a wordâ€™s meaning. Working with such high-dimensional data is complex and computationally expensive.

2. **Reducing Dimensions**: UMAP helps by transforming this high-dimensional data into a lower-dimensional space, typically 2 or 3 dimensions.
   - **Why?**: This makes it much easier to visualize and understand the data. Think of it as turning a 3D object into a 2D drawing â€“ itâ€™s simpler to see the overall shape.

3. **Preserving Relationships**: Even though UMAP reduces the number of dimensions, it tries to keep the important relationships between data points intact.
   - **Example**: If two words are similar in the high-dimensional space, they should still be close to each other in the lower-dimensional space.

##### UMAP in BERTopic

In BERTopic, UMAP plays a crucial role in making the topic modeling process more efficient and effective:

- **Visualization**: By reducing the dimensions of BERT embeddings, UMAP allows us to visualize clusters of topics in a 2D or 3D space. This visualization helps in understanding the relationships and similarities between different topics.
  - **Example**: Imagine plotting the reduced data points on a 2D plane. Clusters of points represent different topics, and the distance between clusters indicates how similar or different the topics are.

- **Clustering**: UMAP-preprocessed data is more manageable for clustering algorithms like HDBSCAN. The reduced dimensionality ensures that the clustering algorithm can efficiently identify meaningful clusters without being bogged down by the complexity of high-dimensional data.
  - **Example**: Consider organizing a library with thousands of books. UMAP helps in categorizing the books into a few main sections (reduced dimensions), making it easier to sort them into specific genres or topics (clustering).

##### Practical Implications

To understand the practical implications, let's look at an example scenario where UMAP is used in BERTopic:

1. **Initial Embeddings**: Start with high-dimensional BERT embeddings for a collection of documents. Each document is represented by a vector with 768 dimensions.

2. **UMAP Transformation**: Apply UMAP to reduce these embeddings to 2 or 3 dimensions. This step simplifies the data while retaining the essential relationships between documents.

3. **Visualization**: Plot the reduced embeddings on a 2D plane. Clusters of points become visible, representing different topics. This visualization provides an intuitive understanding of the data's structure.

4. **Clustering with HDBSCAN**: Use the reduced embeddings for clustering with HDBSCAN. The clustering algorithm can now efficiently group similar documents into topics based on the reduced-dimensional space.

5. **Interpretation**: Analyze the resulting clusters to interpret the topics. The reduced dimensions and clustering make it easier to identify and understand the main themes in the document collection.

UMAP is a critical component in BERTopic, offering several advantages:

- **Enhances Visualization**: Makes it easier to visualize and interpret data by plotting reduced embeddings.

#### 2. Efficient Clustering
**Efficient Clustering** is another significant benefit of using UMAP in BERTopic. High-dimensional data is complex and computationally intensive to process. When UMAP reduces this data to lower dimensions, clustering algorithms like HDBSCAN can work more effectively.

- **Why is this important?**: Lower-dimensional data requires less computational power, which speeds up the clustering process. Moreover, the clustering results tend to be more meaningful because the noise and redundancy in high-dimensional data are minimized.

By simplifying the data, UMAP enables HDBSCAN to identify clusters more accurately. This efficiency is particularly important when working with large datasets, as it ensures faster processing times and more reliable topic modeling results. The reduced complexity also means that the clustering algorithm can focus on the true structure of the data, leading to better-defined and more interpretable topics.

In summary, UMAP enhances BERTopic by making the data easier to visualize and cluster, resulting in more efficient and insightful topic modeling.
